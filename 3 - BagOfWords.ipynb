```python
import pandas as pd
import re
import spacy as spc
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
```

#2 - Datos de entrada

```python
oracion1 = "El gobierno de Baja California y Sempra infraestructura decidieron retomar un plan de concientización entre estudiantes jovenes de todos los niveles de educación sobre los efectos de la actividad humana en el clima y las amenazas que se ciernen sobre la vida futura. #ElVigia"
print(oracion1)
oracion2 = "#LibrosParaElMartes querer El joven marinero Edmundo Dantés, es un hombre honrado y cándido que lleva una vida tranquila. Entre sus planes está casarse con la hermosa Mercedes. Te recomendamos 'El conde de Montecristo'. Mucha actividad de amenazas y venganzas estan relacionadas."
print(oracion2)

clase = {'Clase': ["Periodistico", "Literario"]}
```

#3 - Procesamiento

##Limpieza de datos

```python
def limpieza_datos(texto):
  texto = texto.lower()

  texto = re.sub(r"@\S+", "", texto)  # Eliminar menciones a usuarios
  texto = re.sub("http[s]?\://\S+", "", texto)  # Eliminar enlaces
  texto = re.sub(r"#\S+", "", texto)  # Eliminar hashtags
  texto = re.sub(r"[0-9]", "", texto)  # Eliminar números
  texto = re.sub(r"(\(.*\))|(\[.*\])", "", texto)  # Eliminar paréntesis y corchetes
  texto = re.sub(r"\n", "", texto)  # Eliminar caracteres de nueva línea
  texto = re.sub(r"(http[s]?\://\S+)|([\[\(].*[\)\]])|([#@]\S+)|\n", "", texto)  # Eliminar varios patrones
  texto = re.sub(r"(\.)|(,)", "", texto)  # Eliminar puntos y comas
  texto = re.sub(r"[¡!]", "", texto)  # Eliminar signos de admiración
  texto = re.sub(r"[¿?]", "", texto)  # Eliminar signos de exclamación
  #Extra
  texto = re.sub(r"['']", "", texto)  # Eliminar signos de cita
  return texto

oracion_limpia1 = limpieza_datos(oracion1)
print(oracion_limpia1)
oracion_limpia2 = limpieza_datos(oracion2)
print(oracion_limpia2)
```

##Tokenización


```python
tokens1 = word_tokenize(oracion_limpia1)
print(tokens1)
tokens2 = word_tokenize(oracion_limpia2)
print(tokens2)
```

##StopWords

```python
spanish_stopwords = stopwords.words('spanish')
palabras_filtradas1 = [palabra for palabra in tokens1 if palabra not in spanish_stopwords]
print(palabras_filtradas1)

palabras_filtradas2 = [palabra for palabra in tokens2 if palabra not in spanish_stopwords]
print(palabras_filtradas2)
```

##Lematización

```python
nlp = spc.load("es_core_news_sm")
lema = nlp(" ".join(palabras_filtradas1))
oracion_lematizada1 = " ".join([token.lemma_ for token in lema])
print(oracion_lematizada1)

lema = nlp(" ".join(palabras_filtradas2))
oracion_lematizada2 = " ".join([token.lemma_ for token in lema])
print(oracion_lematizada2)
```

#4 - Bolsa de palabras


```python
vectorizador = CountVectorizer()
vectores1 = vectorizador.fit_transform([oracion_lematizada1])
vocabulario1 = vectorizador.get_feature_names_out()

print(vocabulario1)

vectores2 = vectorizador.fit_transform([oracion_lematizada2])
vocabulario2 = vectorizador.get_feature_names_out()

print(vocabulario2)
```

#5 - DataFrame

```python
print(vocabulario1)
```

```python
print(vectores1)
```

```python
df_bw = pd.DataFrame.sparse.from_spmatrix(vectores1,columns = vocabulario1)
df_bw
```

```python
df1_bw = pd.DataFrame.sparse.from_spmatrix(vectores2, columns = vocabulario2)
df1_bw
```

```python
df_bw_total = pd.concat([df_bw, df1_bw])
df_bw_total
```

```python
df_bw_total = df_bw_total.fillna(0)
```

```python
df_bw_total = df_bw_total.set_axis([0,1])
```
